```python
model = SentenceTransformer("distiluse-base-multilingual-cased-v1")
```

```python
corpus1 =[
    "I love playing football.",
    "I enjoy playing soccer.",
    "She is reading a novel.",
    "She is reading a book.",
    "The weather is sunny today."
]
corpus_emb = model.encode(corpus1)
```

```python
cosine_similarity = util.pytorch_cos_sim(corpus_emb,corpus_emb)
avg_sim = cosine_similarity.mean(dim = 1)
outliner_index = avg_sim.argmin().item()
outliner_index
print(f"outliner sentence : {corpus1[outliner_index]} (AVG Similarity : {avg_sim[outliner_index]:.4f})")
```
